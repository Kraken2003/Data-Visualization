```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'/content/drive/MyDrive/Econometrics /big data/good data.csv')
df
```

```python
sname = df['Series Name'].iloc[0:62].tolist()
scodes = df['Series Code'].iloc[0:62].tolist()
metrics = {sname[i]: scodes[i] for i in range(len(sname))}
cname = df['Country Name'].unique()
ccodes = df['Country Code'].unique()
countries = {cname[i]:ccodes[i] for i in range(len(cname))}
```

```python
def compare_series(country_code1, country_code2, series_code):
    # Filter data for the specified country codes and series code
    country1_data = df[(df['Country Code'] == country_code1) & (df['Series Code'] == series_code)]
    country2_data = df[(df['Country Code'] == country_code2) & (df['Series Code'] == series_code)]

    # Get series name
    series_name = country1_data['Series Name'].iloc[0]  # Assuming series name is the same for both countries

    # Melt the data for easy plotting
    country1_data_melted = country1_data.melt(id_vars=['Country Name'], value_vars=df.columns[4:], var_name='Year', value_name=series_name)
    country2_data_melted = country2_data.melt(id_vars=['Country Name'], value_vars=df.columns[4:], var_name='Year', value_name=series_name)

    plt.figure(figsize=(12, 6))
    sns.set_theme(style = 'darkgrid')

    # Use Seaborn's lineplot function
    sns.lineplot(data=country1_data_melted, x='Year', y=series_name, hue='Country Name',palette=['red'])
    sns.lineplot(data=country2_data_melted, x='Year', y=series_name, hue='Country Name',palette=['blue'])

    plt.title(f'{series_name} Comparison: {country_code1} vs {country_code2}')
    plt.xlabel('Year')
    plt.ylabel(series_name)
    plt.legend(title='Country Name')

    # Customizing x-axis ticks for better readability

    plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))  # Adjust the number of visible ticks


    plt.show()
```

```python
compare_series('IND', 'USA', 'EG.USE.PCAP.KG.OE')
```
